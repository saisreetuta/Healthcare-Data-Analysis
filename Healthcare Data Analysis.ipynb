```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset into a Pandas DataFrame
data = pd.read_csv(r'C:\Users\tutha\OneDrive\Desktop\diabetes.csv')


# Displaying the first few rows of the dataset
data.head()
```

```python
# Checking for the missing values in the dataset
data.isnull().sum()
```

```python
# Showing the basic statistics of the dataset
data.describe()
```

```python
# Plotting Age distribution
sns.histplot(data['Age'], bins=20, kde=True)
plt.title('Age Distribution of Patients')
plt.show()
```

```python
# Plotting Glucose levels by Outcome
sns.boxplot(x='Outcome', y='Glucose', data=data)
plt.title('Glucose Levels vs Diabetes Outcome')
plt.show()
```

```python
# Calculating and ploting the correlation matrix
corr = data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Between Variables')
plt.show()
```
